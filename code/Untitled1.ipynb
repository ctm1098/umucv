```python
%matplotlib inline

import numpy             as np
import cv2               as cv
import matplotlib.pyplot as plt
import skimage.io        as io

img = io.imread('..\images\cube3.png')
plt.imshow(img)
```

```python
def rep(x,n):
    l = [x]
    for i in range(n):
        l.append(x)
    return l

def grid(x,n):
    xs = rep(x,n) #lista de x repetido n veces
    xss = rep(np.hstack(xs),n) 
    return np.vstack(xss)

plt.imshow(grid(img,6))
```
